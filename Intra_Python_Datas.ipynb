```python
from google.colab import files
import io
import pandas as pd

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['company.csv']))
```

Nettoyage de donnees

```python
#retire espas vid
df=df.dropna()
```

```python
#retire :  K ak , epi kou tout bagay anfom, voye done yo nan estil nonb ak vigil
#df['salary']=df['salary'].str.replace(',','').str.replace('K','000').astype(float)

# puisqu'on va manipuler en chaine de caracteres, faisons une convertions
#df['salary'] = df['salary'].astype(str).fillna('')  # pour remplacer aussi les espaces vides. Or on n'en a pas
df.loc[:, 'salary'] = df['salary'].astype(str)   # OK

#Avec .loc  on indexe les vraies donnees

df.loc[:, 'salary'] = df['salary'].str.replace(',', '').str.replace('K', '000').astype(float)
```

```python
# conversion en string
df.loc[:, 'years_at_company'] = df['years_at_company'].astype(str)

#ranje years_at_company
df.loc[:,'years_at_company']=df['years_at_company'].str.replace('years','').astype(int)
```

```python
#rabje genre yo
# Pour chaque remplacement, un appel de la methode  replace()
df.loc[:,'gender']=df['gender'].str.replace('Female','F').str.replace('Male','M')
```

```python
"""
    %d : Jour du mois sous forme de nombre décimal (01 à 31)
    %m : Mois sous forme de nombre décimal (01 à 12)
    %Y : Année avec siècle sous forme de nombre décimal (exemple : 2021)
    %y : Année sans siècle sous forme de nombre décimal (exemple : 21 pour 2021)
    %b : Nom abrégé du mois en anglais (exemple : "Jan", "Feb", "Mar")
"""

#Poum change format date yo
model = ['%d-%m-%Y', '%d-%b-%Y', '%Y-%m-%d','%d-%m-%y']

def konvetiDate(dat):
    for grenn in model:
        try:
            return pd.to_datetime(dat, format=grenn, dayfirst=True) # dayfirst, precision a pandas que le premier est jour, puis mois...
        except ValueError:
            continue
    return pd.NaT

df.loc[:, 'hired_date']= df['hired_date'].apply(konvetiDate)
```

```python
# Repatisyon chak anplwaye nan chak depatman.
# retour d'une sorte de dictionnaire...
# suivant le nombre de fois qu'un departement repete, on a le nombre d'employes
#employer_par_departement = df['department'].value_counts()
repartition = pd.crosstab(index=df['employee_id'], columns=df['department'])
repartition = repartition.replace(0, '').replace(1, 'V')
repartition
#print(employer_par_departement)
```
